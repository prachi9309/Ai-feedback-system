```python
from google.api_core.exceptions import ResourceExhausted
LLM_ENABLED = True

def call_llm(prompt):
    global LLM_ENABLED

    if not LLM_ENABLED:
        return '{"predicted_stars": 3, "explanation": "quota fallback"}'

    try:
        response = model.generate_content(prompt)
        return response.text

    except ResourceExhausted:
        print("Quota exhausted â†’ switching to fallback mode")
        LLM_ENABLED = False   # ðŸ”¥ disable LLM forever
        return '{"predicted_stars": 3, "explanation": "quota fallback"}'
```

```python
def prompt_v1(review):
    prompt = f"""Predict the yelp star rating from 1 to 5.
    Review:"{review}"
    return JSON only:
    {{"predicted_stars":1-5,"explanation":"short reason"}}
    """
    return call_llm(prompt)
```

```python
def prompt_v2(review):
    prompt = f"""
    Classify the review using these rules:
    1 star = very bad experience
    3 stars = average experience
    5 stars = excellent experience

    Review:
    "{review}"

    Respond only in JSON:
    {{
      "predicted_stars": 1-5,
      "explanation": "brief reason"
    }}
    """
    return call_llm(prompt)
```

```python
def prompt_v3(review):
    prompt = f"""
    First determine sentiment (negative, neutral, positive),
    then map it to a rating from 1 to 5.
    Do not show steps.

    Review:
    "{review}"

    Return JSON only:
    {{
      "predicted_stars": 1-5,
      "explanation": "short reasoning"
    }}
    """
    return call_llm(prompt)
```

```python
def extract_rating(response):
    try:
        json_text = response.strip().split("{", 1)[1]
        json_text = "{" + json_text
        data = json.loads(json_text)
        return int(data["predicted_stars"])
    except:
        return None
```

```python
from sklearn.metrics import accuracy_score

def evaluate_prompt(prompt_func):
    preds = []
    valid_json = 0

    for review in df_sample['text']:
        response = prompt_func(review)
        rating = extract_rating(response)

        if rating is not None:
            valid_json += 1
        else:
            rating = 3  # fallback

        preds.append(rating)

    accuracy = accuracy_score(df_sample['stars'], preds)
    json_rate = valid_json / len(df_sample)

    return accuracy, json_rate
```

```python
results = []

for name, func in [
    ("Zero-Shot", prompt_v1),
    ("Anchored", prompt_v2),
    ("Reasoning-Guided", prompt_v3)
]:
    acc, json_rate = evaluate_prompt(func)
    results.append([name, acc, json_rate])

results_df = pd.DataFrame(
    results,
    columns=["Prompt Type", "Accuracy", "JSON Validity"]
)

print(results_df)
```
